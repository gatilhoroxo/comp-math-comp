# Fixed-Point Iteration

This notebook implements the **fixed-point iteration algorithm**.

Given a function g(x), the method computes the sequence:

x_{k+1} = g(x_k)

starting from an initial value x0.

In this example we use:

g(x) = 1 - x^3

with initial value:

x0 = 0.5

The iteration is repeated **12 times** and the results are displayed in an **iteration table**.

This notebook is designed to run in **Google Colab**.

```python
import pandas as pd

def fixed_point(g, x0, n_iter):
    """
    Fixed-point iteration.

    Parameters
    ----------
    g : function
        Function defining the iteration x_{k+1} = g(x_k)
    x0 : float
        Initial value
    n_iter : int
        Number of iterations

    Returns
    -------
    table : pandas DataFrame
        Iteration table
    """

    rows = []
    x = x0

    rows.append({"iteration": 0, "x_k": x})

    for k in range(1, n_iter + 1):
        x = g(x)
        rows.append({"iteration": k, "x_k": x})

    table = pd.DataFrame(rows)
    return table
```

```python
# Define the function g(x)
def g(x):
    return 1 - x**3

# Initial value
x0 = 0.5

# Number of iterations
n = 12

# Run fixed-point iteration
table = fixed_point(g, x0, n)

print("Initial value:", x0)
print("Iterations:", n)

# Display iteration table
table
```
